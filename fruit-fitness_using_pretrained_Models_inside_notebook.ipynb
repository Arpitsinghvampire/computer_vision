```python
import cv2
#we need to standardize the image size 

def standardize_shape(img_path):
    z=cv2.imread(img_path)
    resized_image=cv2.resize(z,(224,224))
    
    return resized_image
```

```python
#now lets import the necessary dependencies 
import tensorflow as tf
from tensorflow import keras 
from tensorflow.keras import layers
from tensorflow.keras.layers import *
from keras.utils import plot_model
from keras.models import Model
#here we will try to make two neural networks 
#to one we will  feed the good images and the other network would have the bad images 
#the third model would have the type of the fruit 
```

```python
input1=Input((224,224,3))
conv_1=Conv2D(128,1,padding='same',activation='relu')(input1)
conv_2=Conv2D(128,1,padding='same',activation='relu')(conv_1)
conv_3=Conv2D(64,1,padding='same',activation='relu')(conv_2)
batchnormalization=BatchNormalization()(conv_3)
conv_4=Conv2D(64,1,padding='same',activation='relu')(batchnormalization)
batchnormalization1=BatchNormalization()(conv_4)
conv_5=Conv2D(32,1,padding='same',activation='relu')(batchnormalization1)

conv_8=Conv2D(8,1,padding='same',activation='relu')(conv_5)

flatten11=Flatten()(conv_8)
dense_1=Dense(128,activation='relu')(flatten11)
dropout=Dropout(0.2)(dense_1)
dense_2=Dense(64,activation='relu')(dropout)
dropout1=Dropout(0.3)(dense_2)
dense_3=Dense(32,activation='relu')(dropout1)


dense_4=Dense(28,activation='relu')(dense_3)

dense_5=Dense(10,activation='relu')(dense_4)

output=Dense(2,activation='softmax')(dense_5)


good_model=Model(inputs=input1,outputs=output)

good_model.summary()
```

```python
plot_model(good_model,show_layer_names=True,show_shapes=True)
```

```python
#similarly we do the same for the bad samples
input2=Input((224,224,3))

conv_1=Conv2D(128,1,padding='same',activation='relu',name='conv_1')(input2)
conv_2=Conv2D(128,1,padding='same',activation='relu',name='conv_2')(conv_1)
conv_3=Conv2D(64,1,padding='same',activation='relu',name='conv_3')(conv_2)
batchnormalization=BatchNormalization(name='batchnormalization')(conv_3)

conv_6=Conv2D(32,1,padding='same',activation='relu',name='conv_6')(conv_3)
conv_7=Conv2D(16,1,padding='same',activation='relu',name='conv_7')(conv_6)
batchnormalization3=BatchNormalization(name='batchnormalization3')(conv_7)
conv_8=Conv2D(8,1,padding='same',activation='relu',name='conv_8')(batchnormalization3)

flatten1=Flatten(name='flatten')(conv_8)
dense_1=Dense(128,activation='relu',name='dense_1')(flatten1)
dropout=Dropout(0.2)(dense_1)
dense_2=Dense(64,activation='relu',name='dense_2')(dropout)
dropout1=Dropout(0.3)(dense_2)
dense_3=Dense(32,activation='relu',name='dense_3')(dropout1)

dense_4=Dense(28,activation='relu',name='dense_4')(dense_3)

dense_5=Dense(10,activation='relu',name='dense_5')(dense_4)

output=Dense(2,activation='softmax')(dense_5)

bad_model=Model(inputs=input2,outputs=output)

bad_model.summary()
```

```python
plot_model(bad_model,show_layer_names=True,show_shapes=True)
```

```python
#now lets prepare our x data 
x_image_good=[]
for i in range(len(x_good_train)):
    z=standardize_shape(x_good_train[i])
    x_image_good.append(z)
    if i%1000==0:
        print(f"{i} done ")
    
x_image_bad=[]
for i in range(len(x_bad_train)):
    z=standardize_shape(x_bad_train[i])
    x_image_bad.append(z)
    if i%1000==0:
        print(f"{i} done ")
    
#we have the x labels for the two models 
```

```python
y_test=y_good_test+y_bad_test

x_final_test=[]

for i in range(len(x_good_test)):
    z=standardize_shape(x_good_test[i])
    x_final_test.append(z)
    
for i in range(len(x_bad_test)):
    z=standardize_shape(x_bad_test[i])
    x_final_test.append(z)
    
    
```
