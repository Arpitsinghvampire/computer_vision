```python
#here we will import all the dependencies
from IPython.display import Image,display
import tensorflow as td
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
import numpy as np
import os
import PIL
from tensorflow.keras.layers import Flatten,Input,BatchNormalization,Conv2D,MaxPool2D,Dense,Dropout
from tensorflow.keras.models import Sequential
```

```python
#here we will generate the VGG 16 model
#
class VGG16_model():
  def __init__(self):
    self.input_size=(224,224,3)
    self.model=Sequential()
    self.activation='relu'
    self.kernel_size=(3,3)
    self.padding='same'

  def convolutional_blocks(self,num_filters):
    self.model.add(Conv2D(num_filters,self.kernel_size,padding=self.padding,activation=self.activation))

  def max_pooling(self):
    self.model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))

  #now lets create the fully functional model
  def model_creation(self):
    self.model.add(Conv2D(input_shape=self.input_size,filters=64,kernel_size=self.kernel_size,padding=self.padding,activation=self.activation))
    self.convolutional_blocks(64)
    self.max_pooling()
    self.convolutional_blocks(128)
    self.convolutional_blocks(128)
    self.max_pooling()
    self.convolutional_blocks(256)
    self.convolutional_blocks(256)
    self.convolutional_blocks(256)
    self.max_pooling()
    self.convolutional_blocks(512)
    self.convolutional_blocks(512)
    self.convolutional_blocks(512)
    self.max_pooling()
    self.convolutional_blocks(512)
    self.convolutional_blocks(512)
    self.convolutional_blocks(512)
    self.max_pooling()
    self.model.add(Flatten())
    self.model.add(Dense(4096,activation=self.activation))
    self.model.add(Dense(4096,activation=self.activation))
    self.model.add(Dense(2,activation='softmax'))

  def model_summary(self):
    self.model.summary()

  def plot_model(self):
    keras.utils.plot_model(self.model,show_layer_names=True,show_shapes=True,to_file='model.png')
    display(Image('model.png'))

  def compile_model(self):
    self.model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])




```

```python
#now lets create the instance of the model
vgg16=VGG16_model()

```

```python
#not lets build the model
vgg16.model_creation()
```

```python
#lets print the summary of the model
vgg16.model_summary()
```

```python
#now lets plot the model
vgg16.plot_model()
```

```python
#now in the above part we made the vgg16 architecture ,
#we now make the vgg19 architecture
class VGG19_model():
  def __init__(self):
    self.activation='relu'
    self.input_size=(224,224,3)
    self.model=Sequential()
    self.kernel_size=(3,3)
    self.padding='same'

  def convolutional_blocks(self,num_filters):
    self.model.add(Conv2D(num_filters,self.kernel_size,padding=self.padding,activation=self.activation))

  def maxpooling(self):
    self.model.add(MaxPool2D())

  def model_creation(self):
    self.model.add(Conv2D(64,self.kernel_size,strides=3,input_shape=self.input_size,activation=self.activation,padding=self.padding))
    self.convolutional_blocks(64)
    self.maxpooling()
    self.convolutional_blocks(128)
    self.convolutional_blocks(128)
    self.maxpooling()
    self.convolutional_blocks(256)
    self.convolutional_blocks(256)
    self.convolutional_blocks(256)
    self.convolutional_blocks(256)
    self.maxpooling()
    self.convolutional_blocks(512)
    self.convolutional_blocks(512)
    self.convolutional_blocks(512)
    self.convolutional_blocks(512)
    self.maxpooling()
    self.convolutional_blocks(512)
    self.convolutional_blocks(512)
    self.convolutional_blocks(512)
    self.convolutional_blocks(512)
    self.maxpooling()
    self.model.add(Flatten())
    self.model.add(Dense(4096,activation=self.activation))
    self.model.add(Dense(4096,activation=self.activation))
    self.model.add(Dense(2,activation='softmax'))


  def plots_model(self):
    keras.utils.plot_model(self.model,show_layer_names=True,show_shapes=True,to_file='model2.png')
    display(Image('model2.png'))

  def model_summary(self):
    self.model.summary()

  def compile_model(self):
    self.model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

```

```python
#now lets create an instance of the model
vgg19=VGG19_model()
```

```python
#here we will create the model
vgg19.model_creation()
```

```python
#now lets plot the summary of the model
vgg19.model_summary()
```

```python
#now lets plot the model
vgg19.plots_model()
```
